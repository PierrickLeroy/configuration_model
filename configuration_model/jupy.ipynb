```python
import numpy as np
from module import generate_degreeDistribution, generate_edgeList
import networkx as nx
import matplotlib.pyplot as plt
```

```python
gamma = 2
N = 200
```

```python
deg_distrib = generate_degreeDistribution(gamma, N)
edge_list = generate_edgeList(deg_distrib)
G = nx.from_edgelist(edge_list)
degree_sequence = sorted((d for n, d in G.degree()), reverse=True)
```

```python
# visualizing degree distribution
k, counts = np.unique(np.array(degree_sequence), return_counts=True)
plt.plot(k, counts/counts.sum())
plt.title('Degree Distribution')
plt.xlabel('Degree')
```

```python
plt.title('All components')
nx.draw_networkx(G, with_labels=False, node_size=np.array([d*10 for n, d in G.degree()]))
```

```python
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G0 = G.subgraph(Gcc[0])
nx.draw_networkx(G0, with_labels=False, node_size=np.array([d*10 for n, d in G0.degree()]))
plt.title('Giant component')
```
